## 5. b) For each column with missing values, replace them with the appropriate valuea

```python
age_mean = col_age.mean()
df_training['Age'] = df_training['Age'].fillna(age_mean)

embarked_mode = df_training['Embarked'].mode()[0]
df_training['Embarked'] = df_training['Embarked'].fillna(embarked_mode)
```

```python
import random

# Drop non-null values from the 'Cabin' column and get their indices
non_null_values = df_training['Cabin'].dropna()
missing_value_indices =  df_training[df_training['Cabin'].isna()].index

# Generate random values to replace missing values
random_values = random.choices(non_null_values.tolist(), k=len(missing_value_indices))

# Assign replacement values to missing indices using loc
df_training.loc[missing_value_indices, 'Cabin'] = random_values
```
