```python
# golbal function to remove UNDERSCORE
def func_remove_underscore(strValue):
    value = str(strValue).replace('_', '')

    if value != '':
        return value
    else:
        return np.nan
    
```

```python
# golbal function to convert STRING value to INT or FLOAT else return NaN
def func_convert_string_to_int_or_float(strValue, flag):
    if (strValue == 'nan'):
        return np.nan
    
    if (strValue != '') and (False == pd.isna(strValue)) and (flag == 1):
        strValue = int(strValue)
        if strValue < 0:
            strValue = strValue * -1

        return strValue
    elif (strValue != '') and (False == pd.isna(strValue)) and (flag == 2):
        strValue = float(strValue)
        if strValue < 0.00:
            strValue = strValue * -1

        return strValue
    else:
        return np.nan
```

```python
df['Age'] = df['Age'].apply(func_remove_underscore)
```

```python
df['Age'] = df['Age'].apply(lambda x: func_convert_string_to_int_or_float(x, 1))
```
